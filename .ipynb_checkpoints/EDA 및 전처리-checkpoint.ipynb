 LH 임대주택의 경우, 보증금을 까면서 임대료를 올리는 방법을 이용하여 입주할 수 있다. 특히 몇몇 데이터는 마이홈 주거복지 지도에 공시된 데이터와 LH에서 공시한 초기값과 다르다. 따라서, 최대한 그 간격을 줄여주기 위해 조화평균을 사용하도록 하자.

```python
filled['price'] = (2 * filled['임대보증금'] * filled['임대료']) / (filled['임대보증금'] + filled['임대료']) # 조화평균
filled.head(2)
```

```python
filled['전용면적'] = filled['전용면적'].apply(lambda x: round(x, -1))
regional_median_df = filled.groupby(['지역', '전용면적']).median()
```

```python
regions = regional_median_df.index.get_level_values(0).unique()
fig, axs = plt.subplots(nrows=int(len(regions) / 2), ncols=2, figsize=(15, 5*len(regions)//2))

for i, region in enumerate(regions):
    temp = regional_median_df.loc[region, 'price']
    row = i // 2
    col = i - row * 2
    sns.scatterplot(x=temp.index, y=temp.values, data=temp, ax=axs[row, col])
    axs[row, col].set_title(f"{region} price")
    
plt.show()
```

각 지역구 별 median 값에 대한 Linear Regression 진행하여 회귀식 저장. 이후 missing price가 있는 지역구는 해당 회귀식을 통해 예측하도록 하자. 이게 가능한 이유는 "경험적"으로 전용면적과 price는 인과관계가 존재하는 것을 알고 있고 그 관계 역시 선형적이기 때문이다. (비용의 문제 때문)

```python
from sklearn.linear_model import Lasso # L1 regularazation regression model
from copy import deepcopy

regression_models = {}

for region in regions:
    temp = regional_median_df.loc[region, 'price'].reset_index().to_numpy()
    model = Lasso(alpha=0.9, random_state=3)
    
    model.fit(temp[:, 0].reshape(-1,1), temp[:, 1])
    
    regression_models[region] = deepcopy(model)
    
    del model
```

```python
missing_regions = missing['지역'].unique()

for r in missing_regions:
    model = regression_models[r]
    temp = missing[missing['지역'] == r]
    
    x = temp['전용면적'].values.reshape(-1, 1)
    preds = model.predict(x)
    temp.loc[:, 'price'] = preds
    missing.update(temp)
```

# 이상치 제거

```python
def filter_by_quantile(df, col, lower=0.02, upper=0.95, ):
    df = df.copy(deep=True)
    up = df[col].quantile(upper)
    down = df[col].quantile(lower)
    
    up_df = df[df[col] >= up].copy()
    up_df[col] = up
    
    down_df = df[df[col] <= down].copy()
    down_df[col] = down
    
    df.update(up_df)
    df.update(down_df)
    
    return df
```

```python
cols = merged.corr().columns
print(cols)
filtered = merged.copy(deep=True)

for col in cols:
    filtered = filter_by_quantile(filtered, col)
```

```python
def encode(df, cat_features):
    df = df.copy()
    df['has_commercial'] = df['단지코드'].apply(lambda x: 1 if x in commercial_codes.values else 0)
    df = pd.get_dummies(df, columns=cat_features)
    
    df = df.set_index("단지코드")
    print(df.shape)
    
    return df
```

```python
encoded_merged = encode(merged, cat_features)
encoded_filtered = encode(filtered, cat_features)
encoded_only_apt = encode(only_apt, cat_features)
encoded_outlier_dropped = encode(outlier_dropped, cat_features)
```

```python
from sklearn.model_selection import train_test_split

    
def split_dataset(df, random_state=3):
    train_idx = set(train_df.index.unique()) & set(df.index)
    test_idx = set(test_df.index.unique()) & set(df.index)
    
    train = df.loc[train_idx, :]
    test = df.loc[test_idx, :]
    test = test.drop('등록차량수', axis=1)
    
    y = train.pop('등록차량수')
    X = train.copy()
    
    
    print(X.shape)
    print(y.shape)

    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=random_state)
    return X_train, X_valid, y_train, y_valid, test
```

```python
merged_ds = split_dataset(encoded_merged)
filtered_ds = split_dataset(encoded_filtered)
only_apt_ds = split_dataset(encoded_only_apt)
outlier_dropped_ds = split_dataset(encoded_outlier_dropped)
```

```python
X_train, X_valid, y_train, y_valid, X_test = merged_ds
```

```python
def make_submission(submission_name, model=None, test=None, preds=None):
    
    if preds is None:
        preds = model.predict(test)
        
    sample_sub = pd.read_csv("./data/sample_submission.csv")
    
    test['preds'] = preds
    sub_preds = test.reset_index().groupby("단지코드").mean()['preds']
    test.drop('preds', axis=1, inplace=True)
    
    result = sample_sub['code'].apply(lambda x: sub_preds[x])
    sample_sub['num'] = result
    sample_sub.to_csv(f'./submissions/{submission_name}.csv', index=False)
```

# 아웃라이어 처리한 데이터 셋과의 비교

```python
from xgboost import XGBRegressor, plot_importance
from lightgbm import LGBMRegressor, plot_importance
from sklearn.ensemble import RandomForestRegressor

result = {}

def run_model(ds, model='xgb', save=False):
    X_train, X_valid, y_train, y_valid, X_test = ds
    
    if model=='xgb':
        model = XGBRegressor(n_estimators=10000, **xgb_params)
        model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], eval_metric='mae', early_stopping_rounds=50, verbose=False)
        
    elif model == 'lgbm':
        model = LGBMRegressor(n_estimators=10000, **lgb_params)
        eval_metric=['l1']
        eval_set = [(X_valid, y_valid)]
        model.fit(X=X_train, y=y_train,
            eval_set=eval_set, eval_metric=eval_metric, verbose=0, 
            early_stopping_rounds=50
        )
        
    elif model == 'rf':
        model = RandomForestRegressor(**rf_params)
        model.fit(X=X_train, y=y_train)
    
    result[type(model).__name__] = model.predict(X_test)
    mae = np.sum(abs(model.predict(X_valid) - y_valid)) / len(X_valid)
    
    print(f"{type(model).__name__}'s mae: {mae}")
    
    if save:
        make_submission(f"{type(model).__name__}_{round(mae, 4)}", model=model, test=X_test)
```

```python
models = ['xgb', 'lgbm', 'rf']

for m in models:
    run_model(merged_ds, m)
```

```python
for m in models:
    run_model(filtered_ds, m)
```

```python
for m in models:
    run_model(only_apt_ds, m)
```

```python
for m in models:
    run_model(outlier_dropped_ds, m)
```

 아웃라이어를 처리한 데이터 셋이 동일한 패러미터 하에서 더 좋은 성능을 보여주었다. 따라서, 가공한 데이터셋으로 이후 모델을 예측하도록 한다.
