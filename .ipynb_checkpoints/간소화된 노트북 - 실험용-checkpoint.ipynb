```python
def density(row):
    max_val = max(row['총세대수'], row['단지내주차면수'])
    min_val = min(row['총세대수'], row['단지내주차면수'])
    if max_val - row['등록차량수'] < -300:
        return "crowded"
    
    elif min_val - row['등록차량수'] > 200:
        return "sparse"
    
    else:
        return "normal"

density_added = refined.copy()
density_added['density'] = density_added.apply(density, axis=1)
```

# 모델 데이터셋 가공

```python
from sklearn.preprocessing import LabelEncoder


def encode(df, cat_features, method='one_hot'):
    df = df.copy()
    
    if method == 'one_hot':
        if cat_features:
            df = pd.get_dummies(df, columns=cat_features)
            
    elif method == 'label':
        le = LabelEncoder()
        
        for cat in cat_features:
            df[cat] = le.fit_transform(df[cat])
    
    df = df.set_index("단지코드")
    print(df.shape)
    
    return df
```

```python
from sklearn.model_selection import train_test_split


# baseline 출처: https://rfriend.tistory.com/519
def train_test_split_by_codes(X, y, test_size=0.07, random_state=6):
    np.random.seed(random_state)
    codes = X.index.unique()
    length_codes = len(codes)
    
    test_num = int(length_codes * test_size)
    train_num = length_codes - test_num
    
    train_idx = np.random.choice(length_codes, train_num, replace=False)
    test_idx = np.setdiff1d(range(length_codes), train_idx)
    
    train_codes = codes[train_idx]
    test_codes = codes[test_idx]
    
    assert set(train_codes) & set(test_codes) == set(), "duplicated entries in splitted indicies"
    
    X_train = X.loc[train_codes, :]
    X_test = X.loc[test_codes, :]
    y_train = y[train_codes]
    y_test = y[test_codes]
    
    return X_train, X_test, y_train, y_test

    
```

```python
def split_dataset(df,  random_state=6, return_valid=True):
    train_idx = set(train_df['단지코드'].unique()) & set(df.index)
    test_idx = set(test_df['단지코드'].unique()) & set(df.index)
    
    train = df.loc[train_idx, :]
    test = df.loc[test_idx, :]
    test = test.drop('등록차량수', axis=1)
    
    y = train.pop('등록차량수')
    X = train.copy()
    
    if return_valid:
        X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.07, random_state=random_state)
        print(X_train.shape, end=', ')
        print(X_valid.shape, end=', ')
        print(f"train ids: {X_train.index.unique().shape}, valid ids: {X_valid.index.unique().shape}")
        return X_train, X_valid, y_train, y_valid, test
    
    else:
        return X, y, test
```

```python
def make_submission(submission_name, X_test=None, model=None, test=None, preds=None, col=0):
    
    if preds is None:
        preds = model.predict(test)
        preds = pd.DataFrame(preds, index=X_test.index)
        preds = preds.reset_index(drop=False).groupby('단지코드').mean()
        print(preds.shape)
        
    sample_sub = pd.read_csv("./data/sample_submission.csv")
    sample_sub['num'] =  sample_sub['code'].apply(lambda x: preds.loc[x, col])
    sample_sub.to_csv(f'./submissions/{submission_name}.csv', index=False)
```

```python
def model_test(n, model, ds_list, plot=True):
    ds_results = []
    mae = 0

    for ds in ds_list:
        X_train, X_valid, y_train, y_valid, X_test = data_sets[ds]

#         X_train = pd.concat([X_train, X_valid])
#         y_train = pd.concat([y_train, y_valid])

#         X_train = X_train.sort_index()
#         y_train = y_train.sort_index()
        X_test = X_test.sort_index()

        if model=='xgb':
            _model = XGBRegressor(n_estimators=n, **xgb_params)

        elif model == 'lgbm':
            _model = LGBMRegressor(n_estimators=n, **lgbm_params)

        elif model == 'rf':
            _model = RandomForestRegressor(n_estimators=n, **rf_params)

        xgb_cv = cross_validate(_model, X_train, y_train, scoring='neg_mean_absolute_error',
                       cv=10, n_jobs = -1, return_estimator=True)

        _model = xgb_cv['estimator']

        _test_pred = []
        for m in _model:
            _test_pred.append(m.predict(X_test))

        _test_pred = reduce(lambda x, y: x + y, _test_pred) / len(_test_pred)
        ds_results.append(_test_pred)
        
        val = []
        val_pred = []
        for m in _model:
            pred = m.predict(X_valid)
            val_pred.append(pred)
            val.append(abs(y_valid - pred))

        l_mae = reduce(lambda x, y: x + y, val) / len(val)
        val_pred = reduce(lambda x, y: x + y, val_pred) / len(val_pred)
        
        mae += np.mean(l_mae)      
        
    mae = round(mae, 3)
    print(f"mae: {mae}")
    
            
    if plot:
        plot_df = pd.DataFrame(val_pred, index=X_valid.index).reset_index()
        plot_df['Truth'] = y_valid.values
        plot_df = plot_df.groupby("단지코드").mean()
        plot_df = plot_df.sort_values('Truth')

        fig, ax = plt.subplots(figsize=(8, 5))

        sns.scatterplot(x=plot_df.index, y=plot_df[0], ax=ax, markers='x')
        sns.scatterplot(x=plot_df.index, y=plot_df['Truth'], ax=ax, markers='o')

        ax.legend(["pred", "truth"], loc=0, frameon=True)
        ax.set_title(f"{model}'s (num_boost: {n}, mae: {mae})")
        ax.set_xticks([])
        ax.set_ylabel("value")
        
        plt.show()
                
    preds = reduce(lambda x, y: x + y, ds_results) / len(ds_results)
    
    preds = pd.DataFrame(preds, index=X_test.index).reset_index()
    _preds = preds.groupby("단지코드").median()
    sample_sub = pd.read_csv(DATA_PREFIX + "sample_submission.csv")
    sample_sub['num'] =  sample_sub['code'].apply(lambda x: _preds.loc[x, 0])
    return sample_sub.set_index('code')
```

```python
import copy
from sklearn.model_selection import KFold


def cat_boost(ds, iterations,
              _cv=False,
              plot=True,
              return_test_pred=False,
              drop_cols=None,
              verbose=100,
              save=False,
              params=cat_params,
              add_cat_feat=[],
              n_folds=1,
              ):
    X_train, X_valid, y_train, y_valid, X_test = cat_data_sets[ds]
    _cat_features = cat_features + add_cat_feat
    mae = None

    if drop_cols:
        X_train = X_train.drop(drop_cols, axis=1)
        X_valid = X_valid.drop(drop_cols, axis=1)
        X_test = X_test.drop(drop_cols, axis=1)

    train_pool = Pool(X_train,
                      y_train,
                      cat_features=_cat_features)

    valid_pool = Pool(X_valid,
                      cat_features=_cat_features)

    test_pool = Pool(X_test,
                     cat_features=_cat_features)

    if _cv:
        p = 'True' if plot is True else 'False'
        scores = cv(train_pool,
                    nfold=10,
                    plot=p,
                    verbose=verbose,
                    params=cat_params,
                    iterations=iterations,
                    )

    else:
        model = CatBoostRegressor(iterations=iterations, verbose=verbose, cat_features=_cat_features, **cat_params)
        models = None

        if n_folds == 1:
            model.fit(train_pool, eval_set=[(X_valid, y_valid)])
            val_pred = model.predict(valid_pool)
            mae = np.round(np.mean(np.abs(y_valid.values - val_pred)), 4)

        else:
            print(f"Using folds: {n_folds}...")
            _, models = kfold_run(X_train, y_train,
                                  n_splits=n_folds, iterations=iterations,
                                  verbose=verbose, cat_features=_cat_features,
                                  cat_params=cat_params)
            val_pred, mae = kfold_predict(models, X_valid, y_valid)


        if plot:
            plot_df = pd.DataFrame(val_pred, index=X_valid.index).reset_index()
            plot_df['Truth'] = y_valid.values
            plot_df = plot_df.groupby("단지코드").median()
            plot_df = plot_df.sort_values('Truth')

            diff = np.abs(plot_df['Truth'] - plot_df[0])
            diff = diff.sort_values(ascending=False)
            print("top 5 error codes:")
            print(diff[:5])

            fig, ax = plt.subplots(figsize=(8, 5))

            sns.scatterplot(x=plot_df.index, y=plot_df[0], ax=ax, markers='x')
            sns.scatterplot(x=plot_df.index, y=plot_df['Truth'], ax=ax, markers='o')

            ax.legend(["pred", "truth"], loc=0, frameon=True)
            ax.set_title(f"Catboost's (num_boost: {iterations}, mae: {mae})")
            ax.set_xticks([])
            ax.set_ylabel("value")

            plt.show()

        if return_test_pred:
            if n_folds == 1:
                test_preds = model.predict(test_pool)

            else:
                test_preds, _ = kfold_predict(models, X_test)
                
            preds = pd.DataFrame(test_preds, index=X_test.index).reset_index()
            _preds = preds.groupby("단지코드").median()
            sample_sub = pd.read_csv(DATA_PREFIX + "sample_submission.csv")
            sample_sub = sample_sub[sample_sub['code'].isin(outlier_codes) == False]
            sample_sub['num'] = sample_sub['code'].apply(lambda x: _preds.loc[x, 0])
            sample_sub = sample_sub.set_index('code')

            if save:
                sample_sub.to_csv(f"./submissions/cat_boost_{n_folds}_{ds}_{iterations}_{mae}.csv")

            return sample_sub
```

```python
def kfold_predict(models, X, y=None):
    iterable = True
    mae = None
    
    try:
        iter(models)

    except TypeError as e:
        iterable = False
    
    if iterable:
        test_pred = np.zeros((X.shape[0], len(models)))

        for i, m in enumerate(models):
            test_pred[:, i] = m.predict(X)

        test_pred_mean = np.median(test_pred, axis=1).reshape(-1, 1)
        
        
        if y is not None:
            mae = np.mean(np.abs(test_pred_mean - y.values))
            print(f"MAE: {round(mae, 4)}")

        return test_pred_mean, mae

    else:
        pred = models.predict(X)

        if y is not None:
            mae = np.mean(np.abs(pred - y.values))
            print(f"MAE: {round(mae, 4)}")

        return pred, mae
```

```python
def kfold_run(X_train, y_train,
              n_splits=10,
              random_state=4,
              iterations=1000,
              verbose=1000,
              cat_features=None,
                cat_params = None,

              ):
    kfold = KFold(n_splits=n_splits, random_state=random_state, shuffle=True)
    train_fold_pred = np.zeros((X_train.shape[0], 1))

    X_train,  y_train = X_train.reset_index(drop=True), y_train.reset_index(drop=True)

    models = []

    for i, (t_i, v_i) in enumerate(kfold.split(X_train)):
        print(f"iteration: {i}")
        x_t = X_train.loc[t_i, :]
        y_t = y_train[t_i]
        val = X_train.loc[v_i, :]

        model = CatBoostRegressor(iterations=iterations, verbose=verbose, cat_features=cat_features,
                                  **cat_params)

        model.fit(x_t, y_t)

        models.append(copy.deepcopy(model))
        train_fold_pred[v_i, :] = model.predict(val).reshape(-1, 1)

    return train_fold_pred, models
```
