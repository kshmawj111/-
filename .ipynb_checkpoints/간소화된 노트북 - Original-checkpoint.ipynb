```python
from sklearn.linear_model import Lasso # L1 regularazation regression model
from copy import deepcopy

models = {}

for col in ['임대보증금', '임대료', 'price']:

    regression_models = {}

    for region in regions:
        temp = regional_median_df.loc[region, col].reset_index().to_numpy()[:-2].copy()
        model = Lasso(alpha=2, random_state=3)

        model.fit(temp[:, 0].reshape(-1,1), temp[:, 1])

        regression_models[region] = deepcopy(model)

        del model
        
    models[col] = regression_models
```

```python
for col in ['임대보증금', '임대료', 'price']:

    missing_regions = missing[pd.isna(missing[col])]['지역'].unique()
    regression_models = models[col]
    
    for r in missing_regions:
        model = regression_models[r]
        temp = missing[missing['지역'] == r].copy()

        x = temp['전용면적'].values.reshape(-1, 1)
        preds = model.predict(x)
        temp.loc[:, col] = preds.copy()
        missing.update(temp)
```

```python
commercial_codes = commercial['단지코드']
cat_features = ['임대건물구분', '지역', '공급유형', '자격유형', 'has_commercial']
```

```python
def encode(df, cat_features):
    df = df.copy()
    
    if cat_features:
        df = pd.get_dummies(df, columns=cat_features)
    
    df = df.set_index("단지코드")
    print(df.shape)
    
    return df
```

```python
from sklearn.model_selection import train_test_split

    
def split_dataset(df, random_state=6):
    train_idx = set(train_df['단지코드'].unique()) & set(df.index)
    test_idx = set(test_df['단지코드'].unique()) & set(df.index)
    
    train = df.loc[train_idx, :]
    test = df.loc[test_idx, :]
    test = test.drop('등록차량수', axis=1)
    
    y = train.pop('등록차량수')
    X = train.copy()
    
    print(X.shape, end=', ')
    print(y.shape)

    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.07, random_state=random_state)
    return X_train, X_valid, y_train, y_valid, test
```

```python
encoded_refined = encode(refined, cat_features)
encoded_outlier_dropped = encode(outlier_dropped, cat_features)
encoded_price_dropped = encode(price_dropped, cat_features)
encoded_almost_dropped = encode(almost_dropped, ['has_commercial'])
encoded_selective = encode(selective, cat_features) 
```

```python
refined_ds = split_dataset(encoded_refined)
outlier_dropped_ds = split_dataset(encoded_outlier_dropped)
price_dropped_ds = split_dataset(encoded_price_dropped)
almost_dropped_ds = split_dataset(encoded_almost_dropped)
selective_ds = split_dataset(encoded_selective)
```

```python
def make_submission(submission_name, model=None, test=None, preds=None):
    
    if preds is None:
        preds = model.predict(test)
        
    sample_sub = pd.read_csv("./data/sample_submission.csv")
    sample_sub['num'] =  sample_sub['code'].apply(lambda x: preds[x])
    sample_sub.to_csv(f'./submissions/{submission_name}.csv', index=False)
```

```python
xgb_params = {'booster': 'dart',
         'max_depth': 6, 'learning_rate': 0.1,
         'objective': 'reg:squarederror',
         'sample_type': 'uniform',
          'colsample_bytree': 0.4,
         'rate_drop': 0.2,
          'subsample': 0.4,
         'skip_drop': 0.1,
          'max_bin': 256,
         'tree_method': 'gpu_hist',
        }
```

```python
lgb_params = {"learning_rate": 0.1,
              'max_depth': 4,
              "num_leaves": 32,
              "min_child_samples": 5,
              "boosting_type": "gbdt",
              "subsample": 0.2,
              "colsample_bytree": 0.2,
              "seed": 42,
              'objective': 'regression_l2',
              "n_jobs": -1,
              'reg_lambda': 0.7,
              
    'force_col_wise':True,
              }
```

```python
rf_params= {
    'n_estimators': 1000,
               'max_depth': 16,
               'min_samples_split': 5,
    'random_state' : 25,
    'n_jobs': 8,
    'oob_score': True,
    'max_samples': 0.9,
              
              }
```

```python
from xgboost import XGBRegressor, plot_importance
from lightgbm import LGBMRegressor, plot_importance
from sklearn.ensemble import RandomForestRegressor

result = {}

def run_model(ds, model='xgb', save=False):
    X_train, X_valid, y_train, y_valid, X_test = ds
    
    if model=='xgb':
        model = XGBRegressor(n_estimators=10000, **xgb_params)
        model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], eval_metric='mae', early_stopping_rounds=50, verbose=False)
        
    elif model == 'lgbm':
        model = LGBMRegressor(n_estimators=10000, **lgb_params)
        eval_metric=['l1']
        eval_set = [(X_valid, y_valid)]
        model.fit(X=X_train, y=y_train,
            eval_set=eval_set, eval_metric=eval_metric, verbose=0, 
            early_stopping_rounds=50
        )
        
    elif model == 'rf':
        model = RandomForestRegressor(**rf_params)
        model.fit(X=X_train, y=y_train)
    
    result[type(model).__name__] = model.predict(X_test)
    mae = np.sum(abs(model.predict(X_valid) - y_valid)) / len(X_valid)
    
    print(f"{type(model).__name__}'s mae: {mae}")
    
    if save:
        make_submission(f"{type(model).__name__}_{round(mae, 4)}", model=model, test=X_test)
```

# 모델이 예측을 제대로 못하는 데이터 살펴보기

```python
test_df = test_df.set_index('단지코드')
```

```python
lgbm_params =  {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'regression_l2',
    'metric': 'mae',
    "learning_rate": 0.1,
    "num_leaves": 32,
    "feature_fraction": 0.4,
    "bagging_fraction": 0.4,
    'bagging_freq': 3,
    "max_depth": 5,
    "reg_alpha": 0.8,
    "reg_lambda": 0.6,
    'verbosity': -1,
    #"min_split_gain":0.2,
    "min_child_weight": 5,
    'force_col_wise': True,
    'min_data_in_leaf': 5,
    'feature_pre_filter': False,
                }
```

```python
from functools import reduce

def pred_to_df(preds, df):
    preds = pd.DataFrame(preds)
    preds['code'] = df.index
    grped = preds.groupby('code').mean()
    return grped

def on_test(lgbm_model, X_test):
    preds = lgbm_model.predict(X_test)
    cv_preds = reduce(lambda x, y: x + y, preds) / len(preds)
    cv_preds = pred_to_df(cv_preds, X_test)
    
    try:
        max_in_feat = X_test.apply(
            lambda x: x['총세대수'] if x['총세대수'] >= x['단지내주차면수'] else x['단지내주차면수'],
            axis=1 )
    
    except KeyError:
        max_in_feat = X_test['selected']
        
    diff = abs(cv_preds[0]- max_in_feat)
    diff = diff.reset_index().groupby("index").mean()
    diff = diff[0].sort_values(ascending=False)
    # diff = diff[diff.values > 10]

    result_df = {'단지코드': [], 'LGBM':[], '총세대수': [], '단지내주차면수': []}

    for code in diff.index:
        result_df['단지코드'].append(code)
        result_df['LGBM'].append(round(cv_preds.loc[code, 0], 4))
        result_df['총세대수'].append(test_df.loc[code, '총세대수'][0]) 
        result_df['단지내주차면수'].append(test_df.loc[code, '단지내주차면수'][0])

    result_df = pd.DataFrame(result_df)
    result_df = result_df.set_index('단지코드')
    result_df['차이'] = result_df.apply(lambda x: min(abs(x['LGBM'] - x['총세대수']), abs(x['LGBM'] - x['단지내주차면수'])), axis=1)
    result_df = result_df.sort_values('차이', ascending=False)
    return result_df

def on_valid(lgbm_model, X_valid, y_valid):
    preds = lgbm_model.predict(X_valid)
    cv_preds = reduce(lambda x, y: x + y, preds) / len(preds)
    cv_preds = pred_to_df(cv_preds, X_valid)
    
    diff = abs(cv_preds[0]- y_valid)
    diff = diff.reset_index().groupby("index").mean()
    diff = diff[0].sort_values(ascending=False)
    # diff = diff[diff.values > 30]
    
    result_df = {'단지코드': [], 'LGBM':[], 'Truth':[]}
    y_valid = y_valid.reset_index().groupby("단지코드").mean()

    for code in diff.index:
        result_df['단지코드'].append(code)
        result_df['LGBM'].append(round(cv_preds.loc[code, 0], 4))
        result_df['Truth'].append(y_valid.loc[code, '등록차량수'])

    result_df = pd.DataFrame(result_df)
    result_df = result_df.set_index('단지코드')
    result_df['차이'] = np.abs(result_df['LGBM'] - result_df['Truth'])

    return result_df
```

```python
def run_test(ds, results_valid, results_test, key, n_rounds=200, verboses=-1):
    X_train, X_valid, y_train, y_valid, X_test = ds
    lgtrain = lgb.Dataset(X_train, y_train ,feature_name = "auto")
    
    lgb_cv = lgb.cv(
        params = lgbm_params,
        train_set = lgtrain,
        num_boost_round=n_rounds,
        stratified=False,
        verbose_eval=verboses,
        nfold = 10,
        return_cvbooster=True,
        early_stopping_rounds=50
    )
    
    lgbm_model = lgb_cv['cvbooster']
    del lgb_cv['cvbooster']
    
    results_test[key] = on_test(lgbm_model, X_test)
    results_valid[key] = on_valid(lgbm_model, X_valid, y_valid)
```

```python
def summarize_diff(results_valid):
    results_valid_df = {}
    for k, v in results_valid.items():
        temp = v['차이'].describe()
        temp['MAE'] = np.mean(v['차이'])
        temp['RMSE'] = np.sqrt(np.mean(v['차이']**2))
        results_valid_df[k] = temp

    results_valid_df = pd.DataFrame(results_valid_df)
    return results_valid_df
```

```python
def get_results(data_sets, n_rounds, verbose=True):
    results_valid = {}
    results_test = {}
    
    for k, ds in data_sets.items():
        verbosity = -1
        
        if verbose:
            verbosity = 100
            print(f"{k} valid")
            
        run_test(ds, results_valid, results_test, k, n_rounds=n_rounds, verboses=verbosity)
        
        if verbose:
            print('#'*40)
            
    results_valid_df = summarize_diff(results_valid)
    results_test_df = summarize_diff(results_test)
    
    return results_valid_df, results_test_df
```
